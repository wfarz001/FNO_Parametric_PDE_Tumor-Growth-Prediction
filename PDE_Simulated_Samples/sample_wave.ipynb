```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Generate time values
t = np.linspace(0, 2 * np.pi, 1000)

# Define the signals
low_frequency_signal = np.sin(10 * t)  # Low frequency
high_frequency_signal = np.sin(50 * t)  # High frequency
summed_signal = low_frequency_signal + high_frequency_signal  # Summation

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 8), sharex=True)

# Plot low-frequency signal
axs[0].plot(t, low_frequency_signal, label="Low Frequency", color="blue")
axs[0].set_title("Low Frequency Signal")
axs[0].set_ylabel("Amplitude")
#axs[0].legend(loc="upper right")

# Plot high-frequency signal
axs[1].plot(t, high_frequency_signal, label="High Frequency", color="green")
axs[1].set_title("High Frequency Signal")
axs[1].set_ylabel("Amplitude")
#axs[1].legend(loc="upper right")

# Plot summation of both signals
axs[2].plot(t, summed_signal, label="Summation", color="red")
axs[2].set_title("Summation of Low and High Frequency Signals")
axs[2].set_xlabel("Time")
axs[2].set_ylabel("Amplitude")
#axs[2].legend(loc="upper right")

# Adjust layout
plt.tight_layout()
# Save the figure as PNG
output_filename = f"/home/wfarz001/Fall-2024_Project/sample_wave.png"
plt.savefig(output_filename, dpi=300)
plt.show()
```
